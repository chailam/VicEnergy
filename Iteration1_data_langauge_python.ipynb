# Iteration 1: data cleaning and wrangling

```python
# import library
import pandas as pd
import numpy as np
```

```python
# create list name
list_name=['89085 Ararat Prison','85279 Bairnsdale Airport','89002 Ballarat Aerodrome','81123 Bendigo Airport',
           '90015 Cape Otway Lighthouse','88110 Castlemaine Prison','85072 East Sale','80015 Echuca Aerodrome',
           '84016 Gabo Island Lighthouse','90173 Hamilton Airport','80023 Kerang','88023 Lake Eildon',
           '80091 Kyabram','87031 Laverton Raaf','79028 Longerenong','88109 Mangalore Airport','88043 Maryborough',
           '86282 Melbourne Airport','76031 Mildura Airport','85280 Morwell (Latrobe Valley Airport)','76047 Ouyen (Post Office)',
          '90171 Portland (Cashmore Airport)','82039 Rutherglen Research','82042 Strathbogie','81049 Tatura Inst Sustainable Ag','76064 Walpeup Research','82138 Wangaratta Aero','85096 Wilsons Promontory Lighthouse']
```

```python
#create a list
data_list=[]
```

```python
# read data 
for data in range(len(list_name)):
    temp_df = pd.read_csv(list_name[data]+".csv")
    data_list.append(temp_df)
```

```python
#get staiobn name,stationlon, stationlat, drop columns, create month and year columns
for j in range(len(data_list)):
    data_list[j]['stationName']=data_list[j]['metadata'][0][5:]
    data_list[j].rename(columns={'station':'stationNumber'},inplace=True)
    data_list[j]['stationLon']=data_list[j]['metadata'][2][11:]
    data_list[j]['stationLat']=data_list[j]['metadata'][1][10:]
    data_list[j]=data_list[j].drop(columns=['daily_rain','daily_rain_source','max_temp','max_temp_source','min_temp','min_temp_source','radiation_source','metadata'])
    data_list[j]['month']=''
    data_list[j]['year']=''
    for k in range(len(data_list[j])):
        data_list[j].loc[k,'month']=data_list[j]['YYYY-MM-DD'][k][5:7]
        data_list[j].loc[k,'year']=data_list[j]['YYYY-MM-DD'][k][0:4]
        
   
```

```python
#concat all data
concat_data=pd.concat(data_list, ignore_index=True)
```

```python
# remove header
df=concat_data[['year', 'month', 'radiation', 'stationNumber']].groupby(['year', 'month','stationNumber']).aggregate('mean')
df['index1']=df.index
df.reset_index(level=0,inplace=True)
df['index1']=df.index
df.reset_index(level=0,inplace=True)
df['index1']=df.index
df.reset_index(level=0,inplace=True)
```

```python
#get index list
index_list=list(df[((df.month == '07') &( df.year == '2021') )].index)
```

```python
#drop the row with the index 
f=df.drop(index_list)
```

```python
#drop columns
df=df.drop(columns=['year','index1'])
```

```python
# MJ -kwh
df['radiation']=df['radiation']/3.6
```

```python
#rename column
df=df.rename(columns = {'radiation':'monthRadiation'})
```

```python
# create csv
df.to_csv('StationData.csv', encoding='utf-8', index=False)
```
